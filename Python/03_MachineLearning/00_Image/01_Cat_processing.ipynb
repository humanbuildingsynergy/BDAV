```python
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = "cat_01.jpg"
img = Image.open(image_path)
```

```python
# Convert the image to a numpy array
img_array = np.array(img)

# Get the dimensions of the image
height, width, channels = img_array.shape
```

```python
# Display the image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
```

```python
# Create DataFrames for each color plane
red_df = pd.DataFrame(img_array[:,:,0], columns=[f'col_{i}' for i in range(width)])
green_df = pd.DataFrame(img_array[:,:,1], columns=[f'col_{i}' for i in range(width)])
blue_df = pd.DataFrame(img_array[:,:,2], columns=[f'col_{i}' for i in range(width)])
```

```python
# Optional: Save DataFrames to CSV files
red_df.to_csv("red_channel.csv", index=False)
green_df.to_csv("green_channel.csv", index=False)
blue_df.to_csv("blue_channel.csv", index=False)
print("\nDataFrames have been saved as CSV files: red_channel.csv, green_channel.csv, blue_channel.csv")
```
